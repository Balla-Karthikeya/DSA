```python
def count_char(text):
    count=0
    vowels="aeiouAEIOU"
    for char in text:
        if char in vowels:
            count=count+1
    return count
def main():
    text=input("Enter you text:")
    print("Char count:",count_char(text))
main()
```

```python
def rev_stri(text):
    rev=""
    for char in text:
        rev=char+rev
    return rev    
def main():
    text=input("Enter text:")
    print("reverse of text:",rev_stri(text))
main()
```

```python
#money management systen
#CLI Expenses Tracker
expenses =[]

def show_menu():
    print("Expenses Tracker")
    print("1.Add Expense")
    print("2.View Expenses")
    print("3.Exit")
def add_expenses():
    name = input("Enter expense name: ")
    try:
        amount=float(input("Enter amount: "))
    except ValueError:
        print("Invalid  amount ! please enter a number: ")
        return
    expense={
        "name":name,
        "amount":amount
    }
    expenses.append(expense)
    print("Expense added successfully.")

def view_expenses():
    if not expenses:
        print("No expenses found.")
        return
    print("-----Expenses List-----")
    total=0
    for expense in expenses:
        print(f" {expense['name']}-{expenses['amount']}")
        total += expense["smount"]
    print("Total Expense: {total}")
def main():
    while True:
        show_menu()
        choice=input("Enter a choice")
        if choice =="1":
            add_expenses()
        elif choice =="2":
            view_expenses()
        elif choice =="3":
            print("Existing...Thank you")
        else:
            print("Invalid Choice.Please try again")
if __name__=="__main__":
    main()
```
