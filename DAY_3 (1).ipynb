```python
#counting Even numbers using functions:

def even_count(n):

    count=0

    for i in range(1,n+1):
        if i%2==0:
            count=count+1
    return count
def main():
    n=int(input("Enter a number:"))
    print("Event count:",even_count(n))
main()
```

```python
#Sum of Odd numbers:
def odd_sum(n):
    sum=0
    for i in range(1,n+1):
        if i%2!=0:
            sum=sum+i
    return sum
def main():
    n=int(input("Enter a number:"))
    print("Odd Sum:",odd_sum(n))
main()
```

```python
def count_pos(n):
    count=0
    for i in n:
        if i>=0:
            count=count+1
    return count
def main():
    a=[1,-2,3,5,-6,7,9,-10]
    print("positive number:",count_pos(a))
main()
```
