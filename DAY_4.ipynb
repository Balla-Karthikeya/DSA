```python
# Counting characters in a string

def count_char(text):
    count = 0
    for char in text:
        count += 1
    return count


def main():
    text = input("Enter name: ")
    print("Count char:", count_char(text))


if __name__ == "__main__":
    main()
```

```python
# Counting vowels in a string

def count_char(text):
    count = 0
    vowels = "aeiouAEIOU"

    for char in text:
        if char in vowels:
            count += 1

    return count


def main():
    text = input("Enter name: ")
    print("Vowels count:", count_char(text))


if __name__ == "__main__":
    main()
```
