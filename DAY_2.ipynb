```python
# day 2 
# if/ else / elif
    # making or choosing conditions based on the logical flow of the program
# exception handling try
    # error handling in the program makes the code run properly after handling the error


# ZeroDivisionError
try:
    num=int(input("Enter a number: "))
    result1 = 10/num
    print("Result: ",result1)
except ZeroDivisionError:
    print("You cannot divide by zero")
finally:
    print("Program ended successfully")
```

```python
# value error 
try:
    num=int(input("Enter a number: "))
    result1 = 10/num
    print("Result: ",result1)
except ValueError:
    print("the number is not valid")
finally:
    print("Program ended successfully")
```

```python
# even or odd
def my_func(n):
    if n%2==0:
        print("even")
    else:
        print("odd")
def main():
    n=int(input("enter a num"))
    my_func(n)
main()
```

```python
# max of two numbers
def my_func(a,b):
    if a>b:
        print("a is max")
    else:
        print("b is max")
def main():
    a=int(input("enter a num"))
    b=int(input("enter a num"))
    my_func(a,b)
main()
```

```python
def calculate_fee(course, marks):
    course = course.strip().lower()
    if course == "ai":
        fee = 50000
    elif course == "web":
        fee = 40000
    elif course == "data science":
        fee = 30000
    else:
        return None  
    if marks > 90:
        print("Discount applied: 5000")
        fee -= 5000  
    return fee
def main():
    print("--- Student Fee Calculator ---")
    course = input("Enter course (AI/WEB/DATA SCIENCE): ")
    
    try:
        marks = int(input("Enter marks (1-100): "))
        if marks < 1 or marks > 100:
            print("Invalid marks. Please enter from 1 to 100.")
            return
    except ValueError:
        print("Invalid input, marks must be a number.")
        return

    fee = calculate_fee(course, marks)
    
    if fee is None:
        print("Invalid course selected.")
    else:
        print(f"Final fee: {fee}")

if __name__ == "__main__":
    main()
```
